## 연령별 인구 시각화

```python
import csv
import matplotlib.pyplot as plt


with open('../datas/age_2023.csv', 'r') as f:
    data = csv.reader(f)
    next(data)
    result = [] # 연령별 인구 저장할 리스트

    for row in data:
        if '평창동' in row[0]:
            for i in row[3:]: # 3번 인덱스 ~ 100번 인덱스
                result.append(int(i.replace(",",'')))

print(len(row[3:]))
# print(result)
plt.rc('font', family='Malgun Gothic')
plt.style.use("ggplot")
plt.title("연령별")
plt.plot(result, 'r') # y축값
plt.xlabel('연령')
plt.ylabel('인구')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt


with open('../datas/age_2023.csv', 'r') as f:
    data = csv.reader(f)
    next(data)
    result = [] # 연령별 인구 저장할 리스트
    # 지역의 이름 입력받기
    name = input("지역의 이름(읍면동 단위)을 입력해주세요 : ")
    
    for row in data:
        if '평창동' in row[0]:
            for i in row[3:]: # 3번 인덱스 ~ 100번 인덱스
                result.append(int(i.replace(",",'')))

print(len(row[3:]))
# print(result)
plt.rc('font', family='Malgun Gothic')
plt.style.use("ggplot")
plt.title(name + "지역의 인구 구조")
#plt.plot(result, 'r') # y축값
plt.bar(range(101), result) # 막대 그래프 x축(0-100), y축 - result
plt.xlabel('연령')
plt.ylabel('인구')
plt.show()
```
